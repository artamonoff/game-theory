```python
# Работа с матрицами
import numpy as np
# Для поиска равновесия
import nashpy as nash
import pandas as pd
```

```python
value = np.arange(50, 101)
A = np.full((len(value), len(value)), 0)
B = np.full((len(value), len(value)), 0)
```

```python
for i in range(len(value)):
    for j in range(len(value)):
        if i==j: 
            A[i,j]=value[i]
            B[i,j]=value[i]
        elif value[i]>value[j] :
            A[i,j]=value[j]-2
            B[i,j]=value[j]+2
        elif value[i]<value[j]:
            A[i,j]=value[i]+2
            B[i,j]=value[i]-2
```

```python
A_df = pd.DataFrame(A, index=value, columns=value).astype(str)
B_df = pd.DataFrame(B, index=value, columns=value).astype(str)
A_df+','+B_df
```

```python
# Задаём игру
nzsg = nash.Game(A,B)

nash_eq = nzsg.support_enumeration()

for eq in nash_eq:
    print(eq)
    print(nzsg[eq])
```
